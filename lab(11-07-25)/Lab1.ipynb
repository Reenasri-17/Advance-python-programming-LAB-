```python
import matplotlib.pyplot as plt  # importing packages
x = [1, 2, 3, 4, 5]  # creating instances with arguments
y = [2, 3, 5, 7, 11]
# Creating a line plot
plt.plot(x, y, marker='o') # using built-in functions to plot a graph
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show() 
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# Creating a scatter plot
plt.scatter(x, y, color='red')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show() 
```
