### The Life cycle of a Plot
#### 1. Scatter plot

```python
# 1. Initial Setup --> importing necessary library and packages.
import matplotlib.pyplot as plt
import numpy as np

# 1. Create the Figure and Axes
# The figure is the overall window; the axes is the plotting area
fig, ax = plt.subplots(figsize=(8, 6))
```

```python
#2. Data preparation
# Create some sample data
np.random.seed(42)  # for reproducibility
x_data = np.random.rand(50) * 100
y_data = np.random.rand(50) * 100
```

```python
# 3. Plot the data
# Use ax.scatter() for a scatter plot
ax.scatter(x_data, y_data, c='skyblue', alpha=0.7, edgecolors='k')
```

```python
# 4. Customize and Decorate the plot
ax.set_title('Random Data Scatter Plot')
ax.set_xlabel('X-Axis Values')
ax.set_ylabel('Y-Axis Values')
ax.grid(True, linestyle='--', alpha=0.6)
ax.axhline(y=50, color='r', linestyle='-', linewidth=2, label='Horizontal line at y=50')
ax.legend()
```

```python
# 4. Render the plot
plt.show()

# You can also save the figure to a file
# fig.savefig('scatter_plot.png')
```

### Image -->Change the characteristics of the image

```python
from PIL import Image #importing necessary libraries and packages

import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.asarray(Image.open('bug.png')) #importing images
print(repr(img)) #represents the pixel values of original image 
```

```python
imgplot = plt.imshow(img) 
#This command takes a 2D or 3D array of numerical pixel values (img) and visualizes them by assigning a color to each value based on a colormap.
#For grayscale images (2D array), it maps lower numerical values to darker shades and higher values to lighter shades. By default, Matplotlib uses the Viridis colormap
#For color images (3D array), the three dimensions (often representing Red, Green, and Blue) are interpreted as color channels, and the values are used to directly determine the color of each pixel.
```

```python
#generating gray scale images
if len(img.shape) == 2:
    lum_img = img  # No need to index, the image is already grayscale
    plt.imshow(lum_img, cmap='gray')  # Specify cmap='gray' for proper display
else:
    # If the shape is 3D (e.g., (height, width, 3)), it's a color image.
    # We can then select a single color channel (e.g., the red channel).
    lum_img = img[:, :, 0]
    plt.imshow(lum_img, cmap='gray')
    
plt.title("Grayscale Image")
plt.axis('off')  # Hide axes
plt.show()

#if the image is already grayscale (len(img.shape) == 2), its pixel values remain unchanged
#if the image is a color image (len(img.shape) == 3), the pixel values are transformed by a selective slicing operation. 
#The line lum_img = img[:, :, 0] extracts only the pixel values from the first color channel (index 0, which is typically the red channel in an RGB image) and discards the others (green and blue). 
#This effectively creates a new 2D grayscale image from just the red color channel's data.
```

```python
plt.imshow(lum_img, cmap="hot")
#The "hot" colormap is a sequential colormap that transitions from black to red, then to yellow, and finally to white. This means:
#Low pixel values will be displayed as black ⬛️.
#Intermediate pixel values will transition through red and yellow 🟥🟨.
#High pixel values will be displayed as white ⬜️.
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral') 
#This command changes the colormap used by the imgplot object. 
#It takes the same original pixel values and re-maps them to colors using the new 'nipy_spectral' colormap, which is designed to highlight different data ranges with a broad spectrum of colors.
```

```python
imgplot = plt.imshow(lum_img)
plt.colorbar() #Using plt.colorbar() after plt.imshow() doesn't change the pixel values; it simply adds a scale bar to the plot to help you interpret the displayed colors as numerical values.
#it adds a reference scale (colorbar) beside the plot, which maps the pixel intensity values (numerical data) to their corresponding colormap colors.
#added reference!
```

```python
plt.hist(lum_img.ravel(), bins=range(256), fc='k', ec='k') #Counts how many pixels fall into each intensity level.
```

### Inference : 
    - Huge spike near 0 → Most pixels are very dark/black.

    - Small peak near 255 → Some pixels are very bright/white.

    - Very few pixels in between → Almost no mid-tones (grays).

    - Overall → The image has high contrast: mostly dark areas, with some bright spots, and not much gradual shading.
